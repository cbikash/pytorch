```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train).to(dtype=torch.float32)
X_test_tensor = torch.from_numpy(X_test).to(dtype=torch.float32)
y_train_tensor = torch.from_numpy(y_train).to(dtype=torch.float32)
y_test_tensor = torch.from_numpy(y_test).to(dtype=torch.float32)
```

```python
from torch.utils.data import Dataset, DataLoader

class CustomDatasetNN(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)
    
    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
train_dataset = CustomDatasetNN(X_train_tensor, y_train_tensor)
test_dataset = CustomDatasetNN(X_test_tensor, y_test_tensor)
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=3, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=3, shuffle=True)
```

```python
import torch.nn as nn 

class SimpleNN(nn.Module):

    def __init__(self, num_features):
        super().__init__()

        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        out = self.linear(features)
        out = self.sigmoid(out)

        return out
```

```python
learning_rate = 0.1
epochs = 25
```

```python
model = SimpleNN(X_train_tensor.shape[1])

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
loss_function = nn.BCELoss()


```

```python
for epoch in range(epochs):
    for batch_feature, batch_labels in train_dataloader:
        y_pred = model(batch_feature)

        loss = loss_function(y_pred, batch_labels.view(-1,1))

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()
        
    
```

```python
model.eval() # set the model to evaluation mode

accuracy_list = []

with torch.no_grad():
    for batch_feature, batch_labels in test_dataloader:
        #forward pass
        y_pred = model(batch_feature)
        y_pred = (y_pred > 0.8).float()

        batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
        accuracy_list.append(batch_accuracy)

overall_accuracy = sum(accuracy_list) /len(accuracy_list)
print(f'Accuracy: {overall_accuracy:.4f}')
```
