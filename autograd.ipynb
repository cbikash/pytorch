```python
x = torch.tensor(3.0, requires_grad=True)
```

```python
def binary_cross_entropy_loss(y_pred, y):
    eplison = 1e-7
    prediction = torch.clamp(y_pred, eplison, 1 - eplison)
    return - (y * torch.log(prediction) + (1 - y) * torch.log(1 - prediction))
```

```python
y = x ** 2
z = torch.sin(y)
```

```python
y
```

```python
z
```

```python
z.backward()
```

```python
x.grad
```

```python
x.requires_grad_(False)
```

```python
y = x**2 
y
```

```python
x = torch.tensor(3.0, requires_grad=True)
y = x**2

z = x.detach()
y1 = z**2
```
